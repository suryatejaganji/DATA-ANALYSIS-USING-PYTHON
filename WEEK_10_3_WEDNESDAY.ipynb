```python
import math

# Factorial function
def factorial(n):
    return math.factorial(n)

# Combination function nCr
def nCr(n, r):
    return factorial(n) / (factorial(r) * factorial(n - r))

# Binomial PMF function
def binomial_pmf(n, k, p):
    return nCr(n, k) * (p**k) * ((1 - p)**(n - k))

# Parameters
n = 15
p = 0.3

# Display probability for k = 0 to 15
print("Binomial PMF Values (Custom Function):")
for k in range(n + 1):
    print(f"P(X = {k}) = {binomial_pmf(n, k, p):.4f}")
```

```python
from scipy.stats import binom
import numpy as np

# Probability exactly 5 customers open the email
prob_equal_5 = binom.pmf(5, n, p)

# Probability at most 5 customers open the email
prob_atmost_5 = binom.cdf(5, n, p)

# Smallest x such that P(X ≤ x) = 0.9
x_90 = binom.ppf(0.9, n, p)

# Generate 20 random samples
random_samples = binom.rvs(n, p, size=20)

print("\nScipy Binomial Results:")
print(f"P(X = 5) = {prob_equal_5:.4f}")
print(f"P(X ≤ 5) = {prob_atmost_5:.4f}")
print(f"Smallest x such that P(X ≤ x) = 0.9: {int(x_90)}")
print(f"20 Random Samples: {random_samples}")
```

```python
import matplotlib.pyplot as plt

x = np.arange(0, n + 1)

pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)

# PMF Plot
plt.figure(figsize=(10, 5))
plt.stem(x, pmf, basefmt=" ")
plt.title("Probability of Customers Opening Emails (PMF)")
plt.xlabel("Number of Customers Opening Email")
plt.ylabel("Probability")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# CDF Plot
plt.figure(figsize=(10, 5))
plt.step(x, cdf, where='post')
plt.title("Cumulative Probability of Email Opens (CDF)")
plt.xlabel("Number of Customers Opening Email")
plt.ylabel("Cumulative Probability")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
