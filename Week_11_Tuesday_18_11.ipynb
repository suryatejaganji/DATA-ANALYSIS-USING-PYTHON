```python
from scipy.stats import binom
import numpy as np

# Parameters for the Binomial distribution
n = 60  # Number of participants
p = 0.78 # Probability of developing immunity

# Number of participants for which we want to find the probability
k_exact = 50

# Calculate the probability that exactly k_exact participants develop immunity
probability_exact_50 = binom.pmf(k_exact, n, p)

print(f"The probability that exactly {k_exact} participants develop immunity is: {probability_exact_50:.4f}")
```

```python
from scipy.stats import poisson

# Define the average rate (lambda)
lambda_rate = 5

# Define the number of reports (k)
k = 8

# Calculate the probability of exactly 8 reports using the Poisson PMF
probability_exact_8 = poisson.pmf(k, lambda_rate)

print(f"The probability of exactly {k} side-effect reports with an average rate of {lambda_rate} is: {probability_exact_8:.4f}")
```

```python
import scipy.stats as stats

# Define the maximum number of reports for cumulative probability
k_max = 5

# The average rate (lambda_rate) is already defined as 5 from previous steps
# lambda_rate = 5

# Calculate the probability of 5 or fewer reports using the Poisson CDF
probability_5_or_fewer = stats.poisson.cdf(k_max, lambda_rate)

print(f"The probability of 5 or fewer side-effect reports with an average rate of {lambda_rate} is: {probability_5_or_fewer:.4f}")
```

```python
from scipy.stats import poisson

# The average rate (lambda_rate) is already defined as 5 from previous steps
# lambda_rate = 5

# Define the desired percentile
percentile_q = 0.90

# Calculate the 90th percentile using the Poisson PPF
pct_90 = poisson.ppf(percentile_q, lambda_rate)

print(f"The 90th percentile for daily side-effect reports with an average rate of {lambda_rate} is: {int(pct_90)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# The average rate (lambda_rate) is already defined as 5 from previous steps
# lambda_rate = 5

# Set the number of simulation days
simulation_days = 30

# Generate 30 random samples from a Poisson distribution
simulated_reports = np.random.poisson(lambda_rate, simulation_days)

# Create a histogram of the simulated data
plt.figure(figsize=(10, 6))
plt.hist(simulated_reports, bins=np.arange(min(simulated_reports)-0.5, max(simulated_reports)+1.5, 1), rwidth=0.8, density=False, alpha=0.7, color='skyblue', edgecolor='black')

# Add title and labels
plt.title('Simulated Daily Side-Effect Reports over 30 Days (\u03bb = 5)')
plt.xlabel('Number of Reports')
plt.ylabel('Frequency')
plt.xticks(np.arange(min(simulated_reports), max(simulated_reports)+1, 1))
plt.grid(axis='y', alpha=0.75)

# Display the plot
plt.show()

print(f"Simulated daily reports for {simulation_days} days: {simulated_reports}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# The average rate (lambda_rate) is already defined as 5 from previous steps
# lambda_rate = 5

# Define the range of reports for plotting
reports_range = np.arange(0, 16) # From 0 to 15 reports

# Calculate PMF for each value in the range
pmf_values = poisson.pmf(reports_range, lambda_rate)

# Calculate CDF for each value in the range
cdf_values = poisson.cdf(reports_range, lambda_rate)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot PMF
axes[0].bar(reports_range, pmf_values, color='skyblue', edgecolor='black')
axes[0].set_title(f'Poisson PMF (\u03bb = {lambda_rate})')
axes[0].set_xlabel('Number of Reports')
axes[0].set_ylabel('Probability')
axes[0].set_xticks(reports_range)
axes[0].grid(axis='y', alpha=0.75)

# Plot CDF
axes[1].plot(reports_range, cdf_values, marker='o', linestyle='-', color='salmon')
axes[1].set_title(f'Poisson CDF (\u03bb = {lambda_rate})')
axes[1].set_xlabel('Number of Reports')
axes[1].set_ylabel('Cumulative Probability')
axes[1].set_xticks(reports_range)
axes[1].grid(axis='both', alpha=0.75)

plt.tight_layout()
plt.show()
```
