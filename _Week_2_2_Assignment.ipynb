```python
import numpy as np
import pandas as pd

# 1. Define a list of quarters
quarters = ['Q1', 'Q2', 'Q3']

# 2. Define a list of marketing platforms
marketing_platforms = ['Social Media', 'Email Marketing', 'Paid Ads']

# 3. Generate a 3x3 array of random integers between 100 and 1000
lead_data_array = np.random.randint(100, 1001, size=(len(quarters), len(marketing_platforms)))

print("Generated Lead Data Array:")
print(lead_data_array)
```

```python
lead_generation_df = pd.DataFrame(lead_data_array, index=quarters, columns=marketing_platforms)
print("Lead Generation DataFrame:")
print(lead_generation_df)
```

```python
# 1. Generate a 3x3 array of random integers for cost per lead (e.g., between 50 and 500 INR)
cost_per_lead_array = np.random.randint(50, 501, size=(len(quarters), len(marketing_platforms)))

print("Generated Cost-per-Lead Array (INR):")
print(cost_per_lead_array)

# 2. Create the Cost-per-Lead DataFrame
cost_per_lead_df = pd.DataFrame(cost_per_lead_array, index=quarters, columns=marketing_platforms)

print("\nCost-per-Lead DataFrame (INR):")
print(cost_per_lead_df)
```

```python
total_campaign_spend_df = lead_generation_df * cost_per_lead_df

print("\nTotal Campaign Spend DataFrame (INR):")
print(total_campaign_spend_df)
```

```python
print("\nInformation about Total Campaign Spend DataFrame:")
total_campaign_spend_df.info()
```

```python
print("\nDescriptive Statistics for Total Campaign Spend DataFrame:")
total_campaign_spend_df.describe()
```

```python
# 1. Simulate a 15% growth in lead generation for Q4
q4_lead_generation_df = lead_generation_df * 1.15

# 2. Update the index to reflect 'Q4' if we consider this as a single Q4 projection
# For simplicity, we are assuming this growth is applied across all previous quarters
# and then potentially representing a 'projected Q4' by taking average or sum of these inflated values.
# However, the prompt asks to 'increase the values in the original lead matrix by 15%',
# implying the structure remains similar but values are scaled.
# Let's create a DataFrame that represents the *projected* Q4 leads based on this growth.
# A more robust approach might be to add a new row for Q4, but the current instruction implies scaling existing data.

# Let's consider creating a Q4 specific row by averaging the Q1-Q3 values and then applying growth.
# Or, simpler, just illustrate the scaled 'original' matrix as the basis for Q4 thinking.
# Given the prompt, let's keep the scaled matrix, indicating what 'each quarter's leads would be like with 15% growth'.
# A more direct interpretation of 'simulate expected growth in lead generation for Q4' is to project new Q4 values.
# To align with the 'increase the values in the original lead matrix by 15%' and 'simulate expected growth in lead generation for Q4',
# I will create a new dataframe that has the same structure as the original but with inflated values,
# representing what *each quarter's* lead generation *would be* if it grew by 15% for Q4 consideration.

# Let's name it projected_q4_lead_generation_df to be clear that it's a projection.
projected_q4_lead_generation_df = lead_generation_df * 1.15

print("\nProjected Q4 Lead Generation DataFrame (15% Growth):")
print(projected_q4_lead_generation_df)
```

```python
# Transpose the projected Q4 lead generation DataFrame
transposed_q4_projected_df = projected_q4_lead_generation_df.T

print("\nTransposed Q4 Projected Lead Generation DataFrame (Platforms as rows, Quarters as columns):")
print(transposed_q4_projected_df)
```
